# Método do Cotovelo para encontrar o numero ideal de clusters

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df_nao_matriculado = df_nao_matriculado.dropna()
df_matriculado = df_matriculado.dropna()
```

```python
scaler1 = StandardScaler()
X1_scaled = scaler1.fit_transform(df_nao_matriculado[['Idade', 'Quantas Pessoas Moram na Casa']])

# Cotovelo
wcss1 = []
for i in range(1, 11):
    kmeans1 = KMeans(n_clusters=i, random_state=42)
    kmeans1.fit(X1_scaled)
    wcss1.append(kmeans1.inertia_)

plt.plot(range(1, 11), wcss1, marker='o')
plt.title('Método do Cotovelo - df1')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
scaler2 = StandardScaler()
X2_scaled = scaler2.fit_transform(df_matriculado[['Idade', 'Quantas Pessoas Moram na Casa']])

# Cotovelo
wcss2 = []
for i in range(1, 11):
    kmeans2 = KMeans(n_clusters=i, random_state=42)
    kmeans2.fit(X2_scaled)
    wcss2.append(kmeans2.inertia_)

plt.plot(range(1, 11), wcss2, marker='o')
plt.title('Método do Cotovelo - df2')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show()
```

# Analise desses clusters

```python
# Médias por cluster
media = df_matriculado.groupby('cluster')[['Idade', 'Quantas Pessoas Moram na Casa']].mean()
media['Idade'] = media['Idade'].round(0)
media['Quantas Pessoas Moram na Casa'] = media['Quantas Pessoas Moram na Casa'].round(2)
print(media)

# Desvio padrão (para entender a variação)
print(df_matriculado.groupby('cluster')[['Idade', 'Quantas Pessoas Moram na Casa']].std())

df_matriculado.info()
```

```python
def limpar_valor_monetario(valor):
    if isinstance(valor, str):
        # Remove "R$", espaços, e troca vírgula por nada
        valor = valor.replace("R$", "").replace(".", "").replace(",", ".").strip()
        try:
            return float(valor)
        except:
            return None
    return valor

# Aplicar para as colunas de renda
colunas_renda = ['Renda Individual Mensal', 'Renda familiar mensal aproximada', 'Renda per capita']

for col in colunas_renda:
    df_matriculado[col] = df_matriculado[col].apply(limpar_valor_monetario)

print(df_matriculado[colunas_renda].dtypes)
print(df_matriculado[colunas_renda].head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
for cluster in df_matriculado['cluster'].unique():
    subset = df_matriculado[df_matriculado['cluster'] == cluster]
    plt.scatter(subset['Idade'], subset['Quantas Pessoas Moram na Casa'], label=f'Cluster {cluster}', alpha=0.6)

plt.xlabel('Idade')
plt.ylabel('Quantas Pessoas Moram na Casa')
plt.title('Clusters - df_matriculado')
plt.legend()
plt.show()
```
